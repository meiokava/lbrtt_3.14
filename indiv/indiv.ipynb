<font size="4">Применение различных методов морфологических операций к изображению

```python
import cv2 as cv2
import numpy as np
import random
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("bskt.jpg", 0);
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
```

<font size="4">1. Применение эрозии (уменьшение размера объекта на изображении)

```python
erosion = cv2.erode(image, kernel,iterations = 2)
```

<font size="4">2. Применение дилатации (увеличение размера объекта на изображении)

```python
dilation = cv2.dilate(image,kernel,iterations = 2)
```

<font size="4">3. Применение градиент (разность между расширением и эрозией изображения, используется для выделения границ объектов)

```python
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(221)
plt.imshow(image, cmap="gray")
plt.title("_Оригинал_")

plt.subplot(222)
plt.imshow(erosion, cmap="gray")
plt.title("_Эрозия_")

plt.subplot(223)
plt.imshow(dilation, cmap="gray")
plt.title("_Дилатация_")

plt.subplot(224)
plt.imshow(gradient, cmap="gray")
plt.title("_Градиент_")

plt.show()
```

<font size="4">1. Топ-хэт (разница между исходным изображением и открытием)

```python
top_hat = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)
```

<font size="4">2. Блэк-хэт - разница между закрытием входного изображения и входным изображением.

```python
black_hat = cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(221)
plt.imshow(top_hat, cmap="gray")
plt.title("Top Hat")

plt.subplot(222)
plt.imshow(black_hat, cmap="gray")
plt.title("Black Hat")

plt.show();
```
